```python
#install the needed libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#import the dataset for the project
file_path = r"C:\Users\lilmi\Downloads\bank_marketing.csv"
df = pd.read_csv(file_path, header = 0)
print (df.shape)
#df.shape helps us to know the size and structure of the dataset by printing the number of rows and columns in a bracket (41188,22)
df.head()
#df.head() reveals the first 5 rows of the dataset by default unless the number of rows is specified in () while df.tail() reveals the last 5 rows
```

```python
#splitting/subsetting the bank_marketing dataset stored in the variable 'df' into dataframes & storing it in 3 different variables
#dataframes make it easier to analyze and manipulate data
client = df[["client_id", "age", "job", "marital", "education", "credit_default", "housing", "loan"]]
campaign = df[["client_id", "campaign", "month", "day", "duration", "pdays", "previous", "poutcome", "y"]]
economics = df[["client_id", "emp_var_rate", "cons_price_idx", "euribor3m", "nr_employed"]]
```

The next phase is to clean the dataset which will involve renaming columns and cleaning the data

```python
client.head(2)
```

```python
#renaming columns within the client dataframe
client.rename(columns = {'client_id':'id'}, inplace = True)
client.head(2)
```

```python
#Cleaning Text Data in Client
client['education'] = client['education'].str.replace(".", "_").str.replace("unknown","NaN")
client['job'] = client['job'].str.replace(".","")
#The str.replace() function is meant for replacing substrings within strings and operates on string data
#Although the '.str.replace' comes with a warning, it's preferred over the suggested alternative because we need to replace a character  
#within varied text in the 'education' column; the suggested code is for targeted values
client.head(10)
```

CAMPAIGN

```python
#renaming columns within the campaign dataframe
campaign.rename(columns = {'duration':'contact_duration', 'previous': 'previous_campaign_contacts', 'y': 'campaign_outcome',
                         'poutcome': 'previous_outcome'}, inplace = True)
campaign.head(2)
```

```python
campaign['previous_outcome'] = campaign['previous_outcome'].str.replace("success","1").str.replace("failure","0").str.replace("nonexistent","NaN")
campaign['campaign_outcome'] = campaign['campaign_outcome'].str.replace("success","1").str.replace("failure","0")
campaign.tail(15)
```

```python
# Creating a datetime column, I'll start by capitalizing the month column
campaign["month"] = campaign["month"].str.capitalize()
#Check if 'year' column already exists
if 'year' not in campaign.columns:
    #Add a new column called 'year' with the values set at 2022
    campaign.insert(loc=2, column='year', value="2022")
    
# Convert day column into string
campaign['day'] = campaign['day'].astype(str)
#Creating the last_contact_date column which will be a combination of the year, month and day column in that order
last_contact_date = campaign["year"].astype(str) + "-" + campaign["month"] + "-" + campaign["day"]
last_contact_date = pd.to_datetime(last_contact_date, format="%Y-%b-%d")

# Add the newly created last_contact_date column to the table
campaign['last_contact_date'] = last_contact_date

# Removing redundant data
campaign.drop(columns = ["month", "day", "year"], inplace=True)
campaign.tail(5)
```

```python
# Saving the three dataframes: Client, Economics and Campaign to csv and setting index to False
campaign.to_csv('campaign.csv', index = False)
economics.to_csv('economics.csv', index = False)
client.to_csv('client.csv', index = False)
```

Designing the database and creating tables from the csv files

```python
# Creating the client table and storing it in a variable called client_table
client_table = """
CREATE TABLE client
(
	id SERIAL PRIMARY KEY,
    age INTEGER,
    job TEXT,
    education TEXT,
    marital TEXT,
    credit_default BOOLEAN,
    housing BOOLEAN,
    loan BOOLEAN
);
\copy client from 'client.csv' DELIMITER ',' CSV HEADER
"""
```

```python
# Creating the campaign table and storing it in a variable called campaign_table
campaign_table = """
CREATE TABLE campaign
(
	campaign_id SERIAL PRIMARY KEY,
    client_id SERIAL REFERENCES client (id),
    number_contacts INTEGER,
    contact_duration INTEGER,
    pdays INTEGER,
    previous_campaign_contacts INTEGER,
    previous_outcome BOOLEAN,
    campaign_outcome BOOLEAN,
    last_contact_date DATE
);
\copy campaign from 'campaign.csv' DELIMITER ',' CSV HEADER
"""
```

```python
# Creating the economic table and storing it in a variable called economics_table
economics_table = """
CREATE TABLE economics
(
	client_id SERIAL REFERENCES client (id),
    emp_var_rate FLOAT,
    cons_price_idx FLOAT,
    euribor_three_months FLOAT,
    number_employed FLOAT
);
\copy economics from 'economics.csv' DELIMITER ',' CSV HEADER
"""
```
